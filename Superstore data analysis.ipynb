# 3 Data cleaning 

```python
# checking duplicated
df.duplicated().sum()
```

```python
df=df.drop_duplicates()
```

```python
# checking duplicated
df.duplicated().sum()
```

```python
df.shape
```

```python
# so now we check null values
df.isnull().sum()
```

```python
df.info()
```

```python
df=df.dropna(subset=["Row ID"])
df=df.dropna(subset=["Order ID"])
df=df.dropna(subset=["Customer ID"])
df=df.dropna(subset=["Product ID"])
```

```python
df.shape
```

```python
# for date fill using ffill
```

```python
df['Order Date']=df['Order Date'].ffill()
df['Ship Date']=df['Ship Date'].ffill()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
sales_mean=df['Sales'].mean()
sales_mean
```

```python
sales_median=df['Sales'].median()
sales_median
```

```python
df.describe()
```

```python
df['Sales']=df['Sales'].fillna(df["Sales"].median())
df['Quantity']=df['Quantity'].fillna(df["Quantity"].median()) 
df['Discount']=df['Discount'].fillna(df["Discount"].mean())
df['Profit']=df['Profit'].fillna(df["Profit"].median())
```

```python
df.isnull().sum()
```

 # Data type conversions

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=True)
df[["Order Date", "Ship Date"]].dtypes
```

```python
df['Year'] = df['Order Date'].dt.year.astype(str)
df['Month'] = df['Order Date'].dt.month_name().str[:3]
```

# dataset modification by adding two columns ,year and month

```python
df["Postal Code"] = df["Postal Code"].astype(str)
df["Row ID"] = df["Row ID"].astype(str)
```

```python
df.info()
```

```python
categorical_Columns=["Customer Name","Segment","Country","City","State","Region","Category","Sub-Category","Product Name","Ship Mode"]
for col in categorical_Columns:
    df[col] = df[col].fillna(df[col].mode()[0])
```

# 4.Exploratory data analysis

```python
# Total sales by category
```

```python
sns.barplot(x="Category",y="Sales",data=df,estimator=np.sum,palette=["#d9be68", "#046980", "#7bd620"],hue="Category",legend=False)
plt.title("Sales by category")
plt.show()
```

```python
# Total sales by subcategory
```

```python
sales_subcategory=df.groupby(["Sub-Category"],as_index=False)["Profit"].sum().sort_values(by="Profit",ascending=False)
cmap=plt.get_cmap("autumn")
norms=plt.Normalize(sales_subcategory["Profit"].min(),sales_subcategory["Profit"].max())
colors=cmap(norms(sales_subcategory["Profit"].values)).tolist()
ax=sns.barplot(y='Sub-Category',x="Profit",data=sales_subcategory,palette=colors,hue="Sub-Category",legend=False)
plt.title("Profit by Subcategory")
plt.show()
```

```python
# Total sales by subcategory
```

```python
sales_subcategory=df.groupby(["Sub-Category"],as_index=False)["Sales"].sum().sort_values(by="Sales",ascending=False)
cmap=plt.get_cmap("plasma")
norms=plt.Normalize(sales_subcategory["Sales"].min(),sales_subcategory["Sales"].max())
colors=cmap(norms(sales_subcategory["Sales"].values)).tolist()
ax=sns.barplot(y='Sub-Category',x="Sales",data=sales_subcategory,palette=colors,hue="Sub-Category",legend=False)
plt.title("Sales by Subcategory")
plt.show()
```

```python
# Dicount vs profit analysis
```

```python
colors = np.where(df["Profit"] >= 0, "#0ef04a", "#82082f")
plt.scatter(df["Discount"], df["Profit"],c=colors,s=80,alpha=0.5)
plt.title("Discount vs Profit Analysis")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.show()
```

```python
# each year monthly profit trend analysis
g=sns.FacetGrid(monthly_Profit,col="Year",col_wrap=2,height=4)
g.map(sns.lineplot,"Month","Profit",linestyle="--")
g.set_titles("Year {col_name}")
g.fig.suptitle("Monthly Profit Treend by per year",y=1.05)
plt.show()
```
